```python
import numpy as np
from scipy.ndimage import uniform_filter1d
```

```python
# Define the function to remove outliers based on the standard deviation
def remove_outliers(df, threshold=3):
    # Compute the mean and standard deviation of the spectra
    mean_spectrum = df.iloc[:, 2:-6].mean(axis=0)
    std_spectrum = df.iloc[:, 2:-6].std(axis=0)
    # Filter out rows with values beyond the threshold
    filtered_df = df[(np.abs(df.iloc[:, 2:-6].sub(mean_spectrum, axis=1)) <= threshold * std_spectrum).all(axis=1)]
    return filtered_df

# Apply the outlier removal function
df_cleaned = remove_outliers(df)

# Calculate the average spectrum
average_spectrum = df_cleaned.iloc[:, 2:-6].mean(axis=0)

# Smooth the average spectrum using a uniform filter
smoothed_spectrum = uniform_filter1d(average_spectrum, size=10)

# Set wavenumber labels (assuming the data starts at 4000 and ends at 650)
wavenumbers = np.linspace(4000, 650, num=len(average_spectrum))

# Reverse the average spectrum data to match the axis
average_spectrum_reversed = average_spectrum[::-1]
smoothed_spectrum_reversed = smoothed_spectrum[::-1]

# Plot the average spectrum without smoothing
plt.figure(figsize=(10, 6))
plt.plot(wavenumbers, average_spectrum_reversed)
plt.xlabel('Wavenumber')
plt.ylabel('Intensity')
plt.title('Average Spectrum')
plt.grid(True)
plt.gca().set_facecolor('white')
plt.legend().remove()
plt.xticks(np.arange(4000, 600, -500))  # Customize tick marks to display desired wavenumbers
plt.gca().invert_xaxis()  # Invert the x-axis to correctly display wavenumbers
plt.minorticks_on()
plt.tick_params(axis='x', which='minor', direction='in', length=4)
plt.tick_params(axis='x', which='major', direction='in', length=7)
plt.tick_params(axis='y', which='minor', direction='in', length=4)
plt.tick_params(axis='y', which='major', direction='in', length=7)
plt.show()

# Plot the smoothed average spectrum
plt.figure(figsize=(10, 6))
plt.plot(wavenumbers, smoothed_spectrum_reversed)
plt.xlabel('Wavenumber')
plt.ylabel('Intensity')
plt.title('Smoothed Average Spectrum')
plt.grid(True)
plt.gca().set_facecolor('white')
plt.legend().remove()
plt.xticks(np.arange(4000, 600, -500))  # Customize tick marks to display desired wavenumbers
plt.gca().invert_xaxis()  # Invert the x-axis to correctly display wavenumbers
plt.minorticks_on()
plt.tick_params(axis='x', which='minor', direction='in', length=4)
plt.tick_params(axis='x', which='major', direction='in', length=7)
plt.tick_params(axis='y', which='minor', direction='in', length=4)
plt.tick_params(axis='y', which='major', direction='in', length=7)
plt.show()
```

```python
from scipy.signal import find_peaks
```

```python
# Detect peaks in the smoothed spectrum
peaks, _ = find_peaks(smoothed_spectrum_reversed, height=0)  # Adjust 'height' to detect relevant peaks

# Extract peak positions and intensities
peak_wavenumbers = wavenumbers[peaks]
peak_intensities = smoothed_spectrum_reversed[peaks]

# Create a DataFrame to store peak information
peak_df = pd.DataFrame({
    'Wavenumber': peak_wavenumbers,
    'Intensity': peak_intensities
})

# Display the peaks DataFrame
print(peak_df)

# Plot the smoothed average spectrum with peaks marked
plt.figure(figsize=(10, 6))
plt.plot(wavenumbers, smoothed_spectrum_reversed, label='Smoothed Average Spectrum')
plt.plot(peak_wavenumbers, peak_intensities, 'ro', label='Peaks')  # Mark the peaks with red circles
plt.xlabel('Wavenumber')
plt.ylabel('Intensity')
plt.title('Smoothed Average Spectrum with Peaks')
plt.grid(True)
plt.gca().set_facecolor('white')
plt.legend()
plt.xticks(np.arange(4000, 600, -500))  # Customize tick marks to display desired wavenumbers
plt.gca().invert_xaxis()  # Invert the x-axis to correctly display wavenumbers
plt.minorticks_on()
plt.tick_params(axis='x', which='minor', direction='in', length=4)
plt.tick_params(axis='x', which='major', direction='in', length=7)
plt.tick_params(axis='y', which='minor', direction='in', length=4)
plt.tick_params(axis='y', which='major', direction='in', length=7)
plt.show()

```
