```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Preprocessing Functions

class UV_SNV:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        scaler = StandardScaler()
        return scaler.fit_transform(X)

class UV_MSC:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        mean_spectrum = np.mean(X, axis=0)
        return (X - mean_spectrum) / np.std(X, axis=0)

class UV_1der:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return np.gradient(X, axis=1)

class UV_2der:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return np.gradient(np.gradient(X, axis=1), axis=1)

class Par_SNV:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        scaler = StandardScaler()
        return scaler.fit_transform(X) / np.sqrt(np.var(X, axis=0))

class Par_MSC:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        mean_spectrum = np.mean(X, axis=0)
        return (X - mean_spectrum) / np.sqrt(np.var(X, axis=0))
```
