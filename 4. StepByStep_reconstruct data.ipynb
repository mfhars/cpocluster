```python
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler

# Preprocessing Methods
class UV_SNV:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        scaler = RobustScaler()
        return scaler.fit_transform(X)

class UV_MSC:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        mean_spectrum = np.mean(X, axis=0)
        return (X - mean_spectrum) / np.std(X, axis=0)

class UV_1der:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return np.gradient(X, axis=1)

class UV_2der:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return np.gradient(np.gradient(X, axis=1), axis=1)

class Par_SNV:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        scaler = StandardScaler()
        scaled_X = scaler.fit_transform(X)
        variances = np.var(X, axis=0)
        sqrt_variances = np.sqrt(variances)
        if sqrt_variances.shape[0] != scaled_X.shape[1]:
            sqrt_variances = sqrt_variances[:scaled_X.shape[1]]  # Adjust the shape
        return scaled_X / sqrt_variances

class Par_MSC:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        mean_spectrum = np.mean(X, axis=0)
        variances = np.var(X, axis=0)
        sqrt_variances = np.sqrt(variances)
        if sqrt_variances.shape[0] != X.shape[1]:
            sqrt_variances = sqrt_variances[:X.shape[1]]  # Adjust the shape
        return (X - mean_spectrum) / sqrt_variances

preprocessing_methods = {
    'UV-SNV': UV_SNV(),
    'UV-MSC': UV_MSC(),
    'UV-1-der': UV_1der(),
    'UV-2-der': UV_2der(),
    'Par-SNV': Par_SNV(),
    'Par-MSC': Par_MSC()
}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import silhouette_score
from matplotlib.patches import Ellipse

# Function to plot clusters with ellipses and boundaries
def plot_clusters_with_ellipses_and_boundaries(components, labels, title, country_mapping):
    plt.figure(figsize=(10, 7))
    scatter = sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=labels, palette='viridis', legend='full')
    
    # Add filled confidence ellipses
    unique_labels = np.unique(labels)
    palette = sns.color_palette('viridis', len(unique_labels))
    
    for label, color in zip(unique_labels, palette):
        subset = components[labels == label]
        if subset.shape[0] > 1:
            cov = np.cov(subset, rowvar=False)
            mean = np.mean(subset, axis=0)
            eigvals, eigvecs = np.linalg.eigh(cov)
            order = eigvals.argsort()[::-1]
            eigvals = eigvals[order]
            eigvecs = eigvecs[:, order]
            angle = np.degrees(np.arctan2(*eigvecs[:, 0][::-1]))
            width, height = 2 * np.sqrt(eigvals)
            ellipse = Ellipse(xy=mean, width=1.5 * width, height=1.5 * height, angle=angle, 
                              edgecolor=color, facecolor=color, alpha=0.3, lw=1)
            plt.gca().add_patch(ellipse)
    
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    
    # Add custom legend with country names
    handles, _ = scatter.get_legend_handles_labels()
    plt.legend(handles=handles, labels=[country_mapping[label] for label in unique_labels], title='Country Code')
    plt.show()

# Country mapping
country_mapping = {0: 'Thailand', 1: 'Indonesia', 2: 'Malaysia'}

# PCA without any preprocessing
pipeline_no_preprocessing = Pipeline([
    ('pca', PCA(n_components=2))
])
components_no_preprocessing = pipeline_no_preprocessing.fit_transform(remaining_numeric_data)
plot_clusters_with_ellipses_and_boundaries(components_no_preprocessing, remaining_data['country_code'], 'PCA Only', country_mapping)

silhouette_score_no_preprocessing = silhouette_score(components_no_preprocessing, remaining_data['country_code'])
print(f'Silhouette Score for PCA Only: {silhouette_score_no_preprocessing}')
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
import umap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Ensure 'remaining_numeric_data' is updated without 'country_code' and 'country_char'
# Assuming 'remaining_numeric_data' contains only numeric features for clustering

# Mapping of country codes to names
#country_mapping = {0: 'Thailand', 1: 'Indonesia', 2: 'Malaysia'}

# Function to plot clusters with ellipses and boundaries
def plot_clusters_with_ellipses_and_boundaries(components, labels, title) #, country_mapping):
    plt.figure(figsize=(10, 7))
    scatter = sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=labels, palette='viridis', legend='full')
    
    unique_labels = np.unique(labels)
    palette = sns.color_palette('viridis', len(unique_labels))
    
    for label, color in zip(unique_labels, palette):
        subset = components[labels == label]
        if subset.shape[0] > 1:
            cov = np.cov(subset, rowvar=False)
            mean = np.mean(subset, axis=0)
            eigvals, eigvecs = np.linalg.eigh(cov)
            order = eigvals.argsort()[::-1]
            eigvals = eigvals[order]
            eigvecs = eigvecs[:, order]
            angle = np.degrees(np.arctan2(*eigvecs[:, 0][::-1]))
            width, height = 2 * np.sqrt(eigvals)
            ellipse = Ellipse(xy=mean, width=1.5 * width, height=1.5 * height, angle=angle, 
                              edgecolor=color, facecolor=color, alpha=0.3, lw=1)
            plt.gca().add_patch(ellipse)
    
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.show()

# Perform PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(remaining_numeric_data)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(remaining_numeric_data)

# Perform Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(remaining_numeric_data)

# Perform UMAP
umap_reducer = umap.UMAP(n_components=2, random_state=42)
umap_components = umap_reducer.fit_transform(remaining_numeric_data)

# Perform K-Means followed by UMAP
kmeans_umap_labels = kmeans.fit_predict(umap_components)

# Perform Hierarchical Clustering followed by UMAP
hierarchical_umap_labels = hierarchical.fit_predict(umap_components)

# Plot results
plot_clusters_with_ellipses_and_boundaries(pca_components, kmeans_labels, 'K-Means followed by PCA', country_mapping)
plot_clusters_with_ellipses_and_boundaries(pca_components, hierarchical_labels, 'Hierarchical Clustering followed by PCA', country_mapping)
plot_clusters_with_ellipses_and_boundaries(umap_components, kmeans_umap_labels, 'K-Means followed by UMAP', country_mapping)
plot_clusters_with_ellipses_and_boundaries(umap_components, hierarchical_umap_labels, 'Hierarchical Clustering followed by UMAP', country_mapping)
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
import umap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score
from matplotlib.patches import Ellipse

# Preprocessing Classes
class UV_SNV:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        scaler = RobustScaler()
        return scaler.fit_transform(X)

class UV_MSC:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        mean_spectrum = np.mean(X, axis=0)
        return (X - mean_spectrum) / np.std(X, axis=0)

class UV_1der:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return np.gradient(X, axis=1)

class UV_2der:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return np.gradient(np.gradient(X, axis=1), axis=1)

class Par_MSC:
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        mean_spectrum = np.mean(X, axis=0)
        return (X - mean_spectrum) / np.sqrt(np.var(X, axis=0))

# Function to plot clusters with ellipses and boundaries
def plot_clusters_with_ellipses_and_boundaries(components, labels, title, country_labels):
    plt.figure(figsize=(10, 7))
    scatter = sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=country_labels, palette='viridis', legend='full')
    
    unique_labels = np.unique(labels)
    palette = sns.color_palette('viridis', len(unique_labels))
    
    for label, color in zip(unique_labels, palette):
        subset = components[labels == label]
        if subset.shape[0] > 1:
            cov = np.cov(subset, rowvar=False)
            mean = np.mean(subset, axis=0)
            eigvals, eigvecs = np.linalg.eigh(cov)
            order = eigvals.argsort()[::-1]
            eigvals = eigvals[order]
            eigvecs = eigvecs[:, order]
            angle = np.degrees(np.arctan2(*eigvecs[:, 0][::-1]))
            width, height = 2 * np.sqrt(eigvals)
            ellipse = Ellipse(xy=mean, width=width, height=height, angle=angle, 
                              edgecolor=color, facecolor=color, alpha=0.3, lw=1)
            plt.gca().add_patch(ellipse)
    
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.show()

# Preprocessing Methods
preprocessing_methods = {
    'UV-SNV': UV_SNV(),
    'UV-MSC': UV_MSC(),
    'UV-1-der': UV_1der(),
    'UV-2-der': UV_2der(),
    'Par-MSC': Par_MSC()
}

# Apply each preprocessing method and perform clustering
for method_name, method in preprocessing_methods.items():
    print(f"Applying preprocessing method: {method_name}")
    
    # Apply PCA
    pipeline_pca = Pipeline([
        ('preprocessing', method),
        ('pca', PCA(n_components=2))
    ])
    components_pca = pipeline_pca.fit_transform(remaining_numeric_data)
    
    # Apply K-Means followed by PCA
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans_labels = kmeans.fit_predict(remaining_numeric_data)
    plot_clusters_with_ellipses_and_boundaries(components_pca, kmeans_labels, f'K-Means followed by PCA with {method_name}', remaining_data['country_code'])
    
    # Apply Hierarchical Clustering followed by PCA
    hierarchical = AgglomerativeClustering(n_clusters=3)
    hierarchical_labels = hierarchical.fit_predict(remaining_numeric_data)
    plot_clusters_with_ellipses_and_boundaries(components_pca, hierarchical_labels, f'Hierarchical Clustering followed by PCA with {method_name}', remaining_data['country_code'])
    
    # Apply UMAP
    umap_reducer = umap.UMAP(n_components=2, random_state=42)
    components_umap = umap_reducer.fit_transform(remaining_numeric_data)
    plot_clusters_with_ellipses_and_boundaries(components_umap, kmeans_labels, f'K-Means followed by UMAP with {method_name}', remaining_data['country_code'])
    plot_clusters_with_ellipses_and_boundaries(components_umap, hierarchical_labels, f'Hierarchical Clustering followed by UMAP with {method_name}', remaining_data['country_code'])
    
    # Calculate silhouette scores
    silhouette_pca = silhouette_score(components_pca, kmeans_labels)
    silhouette_hierarchical_pca = silhouette_score(components_pca, hierarchical_labels)
    silhouette_umap = silhouette_score(components_umap, kmeans_labels)
    silhouette_hierarchical_umap = silhouette_score(components_umap, hierarchical_labels)
    
    print(f'Silhouette Score for K-Means followed by PCA with {method_name}: {silhouette_pca}')
    print(f'Silhouette Score for Hierarchical Clustering followed by PCA with {method_name}: {silhouette_hierarchical_pca}')
    print(f'Silhouette Score for K-Means followed by UMAP with {method_name}: {silhouette_umap}')
    print(f'Silhouette Score for Hierarchical Clustering followed by UMAP with {method_name}: {silhouette_hierarchical_umap}')

    # Combine clustering results with original country labels
    results = remaining_data[['country_code', 'country_char']].copy()
    results['PCA1'] = components_pca[:, 0]
    results['PCA2'] = components_pca[:, 1]
    results['KMeans_Labels'] = kmeans_labels
    results['Hierarchical_Labels'] = hierarchical_labels
    results['UMAP1'] = components_umap[:, 0]
    results['UMAP2'] = components_umap[:, 1]

    # Display the first few rows of the results
    print(results.head())

    # Evaluate K-Means clustering with PCA
    kmeans_pca_table = results.groupby(['KMeans_Labels', 'country_char']).size().unstack(fill_value=0)
    print(f'K-Means followed by PCA with {method_name}:')
    print(kmeans_pca_table)

    # Evaluate Hierarchical Clustering with PCA
    hierarchical_pca_table = results.groupby(['Hierarchical_Labels', 'country_char']).size().unstack(fill_value=0)
    print(f'Hierarchical Clustering followed by PCA with {method_name}:')
    print(hierarchical_pca_table)

    # Evaluate K-Means with UMAP
    kmeans_umap_table = results.groupby(['KMeans_Labels', 'country_char']).size().unstack(fill_value=0)
    print(f'K-Means followed by UMAP with {method_name}:')
    print(kmeans_umap_table)

    # Evaluate Hierarchical Clustering with UMAP
    hierarchical_umap_table = results.groupby(['Hierarchical_Labels', 'country_char']).size().unstack(fill_value=0)
    print(f'Hierarchical Clustering followed by UMAP with {method_name}:')
    print(hierarchical_umap_table)
```
